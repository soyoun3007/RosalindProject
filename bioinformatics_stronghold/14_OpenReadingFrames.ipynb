```python
from Bio import SeqIO

file = "sample.txt"

genes = list(SeqIO.parse(file, "fasta"))
dna_seq = ""
rna_seq1 = ""
rna_seq2 = ""

for i in range(0, len(genes)):
    dna_seq = str(genes[i].seq)

for n in dna_seq:
    rna_seq1 = dna_seq.replace("T", "U") 

def rev_comp(dna):
    complement = dict(zip("AUGC", "UACG"))
    new_dna = ""
    for l in dna:
        new_dna += complement[l]
    return new_dna[::-1]

rna_seq2 = rev_comp(rna_seq1)

def find_orf(rna1, rna2):
    rna_codon_table = {
        'UUU': 'F',     'CUU': 'L',     'AUU': 'I',     'GUU': 'V',
        'UUC': 'F',     'CUC': 'L',     'AUC': 'I',     'GUC': 'V',
        'UUA': 'L',     'CUA': 'L',     'AUA': 'I',     'GUA': 'V',
        'UUG': 'L',     'CUG': 'L',     'AUG': 'M',     'GUG': 'V',
        'UCU': 'S',     'CCU': 'P',     'ACU': 'T',     'GCU': 'A',
        'UCC': 'S',     'CCC': 'P',     'ACC': 'T',     'GCC': 'A',
        'UCA': 'S',     'CCA': 'P',     'ACA': 'T',     'GCA': 'A',
        'UCG': 'S',     'CCG': 'P',     'ACG': 'T',     'GCG': 'A',
        'UAU': 'Y',     'CAU': 'H',     'AAU': 'N',     'GAU': 'D',
        'UAC': 'Y',     'CAC': 'H',     'AAC': 'N',     'GAC': 'D',
        'UAA': 'Stop',  'CAA': 'Q',     'AAA': 'K',     'GAA': 'E',
        'UAG': 'Stop',  'CAG': 'Q',     'AAG': 'K',     'GAG': 'E',
        'UGU': 'C',     'CGU': 'R',     'AGU': 'S',     'GGU': 'G',
        'UGC': 'C',     'CGC': 'R',     'AGC': 'S',     'GGC': 'G',
        'UGA': 'Stop',  'CGA': 'R',     'AGA': 'R',     'GGA': 'G',
        'UGG': 'W',     'CGG': 'R',     'AGG': 'R',     'GGG': 'G'
    }

    strands = [rna1, rna2]
    orf_seq = []
    orf_list = []
    
    for strand in strands:
        for i in range(0, len(strand)):
            if strand[i:i+3] == "AUG":
                new_strand = strand[i:-1]
                for j in range(0, len(new_strand)-3, 3):
                    if new_strand[j:j+3] in ('UAA', 'UAG', 'UGA'):
                        scan_seq = new_strand[0:j+3]
                        orf_seq.append(scan_seq)
                        break

    for i in range(0, len(orf_seq)):
        aa = ""
        scan_seq = orf_seq[i]
        for j in range(0, len(scan_seq), 3):
            codon = scan_seq[j:j+3]
            tRNA_aa = rna_codon_table[codon]
            if tRNA_aa != "Stop":
                aa += tRNA_aa
        if aa not in orf_list:
            orf_list.append(aa)

    return orf_list

result = find_orf(rna_seq1, rna_seq2)

for n in result:
    print(n)


```
