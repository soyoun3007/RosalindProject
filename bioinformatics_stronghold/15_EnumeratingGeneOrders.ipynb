```python
from itertools import permutations

file = "sample.txt"
with open(file) as f:
    perm_length = int(f.read())
    
    perms = permutations(range(1, perm_length+1), perm_length)
    perms_list = [perm for perm in perms]
    print(len(perms_list))
    for perm in perms_list:
        print(*perm)
```
